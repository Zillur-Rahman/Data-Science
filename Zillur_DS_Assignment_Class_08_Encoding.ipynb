# Ordinal

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df_ord_1 = df.copy() # for Encoding Manually
df_ord_2 = df.copy() # for Encoding using For Loop
```

### Encoding Manually 

```python
df_ord_1.head()
```

```python
unique_region_values = df_ord_1.region.unique()
unique_region_values
```

```python
ordinal = OrdinalEncoder(categories=[unique_region_values])
ordinal
```

```python
encoded = ordinal.fit_transform(df_ord_1[['region']])
```

```python
encoded
```

```python
df_ord_1['region'] = pd.DataFrame(encoded, columns=['region'])
```

```python
df_ord_1.head() # Check the data showing top 5 rows
```

```python
pd.DataFrame(encoded,columns=['region'])
```

### Encoding using For Loop

```python
df_ord_2.head()
```

```python
for col in df_ord_2.columns:
    if is_numeric_dtype(df_ord_2[col]):
        continue
    
    # 1. Get the Unique Values
    unique_values = df_ord_2[col].unique()
    
    # 2. Creating the Ordinal Object using the Unique Values
    ordinal = OrdinalEncoder(categories=[unique_values])
    
    # 3. Encoding
    encoded = ordinal.fit_transform(df_ord_2[[col]])
    
    # 4. Replace the earlier value with the Encoded values
    df_ord_2[col] = pd.DataFrame(encoded, columns=[col])
```
